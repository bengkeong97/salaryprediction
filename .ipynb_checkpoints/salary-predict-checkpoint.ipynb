```python
def shorten_categories (categories, cutoff):
    categorical_map ={}
    
#      loop the 53 country value 
    for i in range (len(categories)):
        if categories.values[i] >= cutoff:
            # store it into dictionary
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            #store it into dictionary with "others"
            categorical_map[categories.index[i]] ='Other'
    return categorical_map
```

```python
categorical_map = shorten_categories(df.Country.value_counts(), 80)
categorical_map
```

```python
df['Country']= df['Country'].map(categorical_map)
df.Country.value_counts()
```

```python
# create a function to clean the string value in the yearcodepro

def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.
    return float(x)
    
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)




```

```python
df['YearsCodePro'].unique()
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post_grad'
    return 'Less than a Bachelors'


```

```python
df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df
```

```python
# model value conflict as Edlvel is string and yearcode pro is float 
#so need to encode the value 

from sklearn import preprocessing
le_education = preprocessing.LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df['EdLevel'].unique()





```

```python
le_country = preprocessing.LabelEncoder()
df['Country'] = le_education.fit_transform(df['Country'])
df['Country'].unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
#model
from sklearn.linear_model import LinearRegression 
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
y_pred = linear_reg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np 
error = np.sqrt(mean_squared_error(y_test, y_pred))
```
